# Relational graph

## line graph

```python


def plot_revenue_over_time(dates, revenues):
    """
    Plots total revenue over time.
    
    Parameters:
    dates (list): A list of date strings representing the dates of revenue data points.
    revenues (list): A list of floats representing the total revenue corresponding to each date.
    
    Returns:
    None
    """
    plt.figure(figsize=(10, 6))
    plt.plot(dates, revenues, marker='o', linestyle='-')
    plt.title('Total Revenue Over Time')
    plt.xlabel('Date')
    plt.ylabel('Total Revenue')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Example data (you would replace this with your actual data)
revenue_dates = ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05']
total_revenues = [10000, 12000, 9000, 15000, 13000]

# Call the function to plot the revenue data
plot_revenue_over_time(revenue_dates, total_revenues)
```

## scatter graph

```python

def plot_sales_vs_revenue(sales_counts, total_revenues):
    """
    Plots total revenue against sales count.
    
    Parameters:
    sales_counts (list): A list of integers representing the sales count for each data point.
    total_revenues (list): A list of floats representing the total revenue corresponding to each sales count.
    
    Returns:
    None
    """
    plt.figure(figsize=(10, 6))
    plt.scatter(sales_counts, total_revenues, color='blue', alpha=0.5)
    plt.title('Total Revenue vs. Sales Count')
    plt.xlabel('Sales Count')
    plt.ylabel('Total Revenue')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Example data (you would replace this with your actual data)
sales_counts = [100, 200, 300, 400, 500]
total_revenues = [10000, 15000, 20000, 25000, 30000]

# Call the function to plot the scatter graph
plot_sales_vs_revenue(sales_counts, total_revenues)
```

## histogram 

```python

def plot_revenue_histogram(total_revenues):
    """
    Plots a histogram of total revenue distribution.
    
    Parameters:
    total_revenues (list): A list of floats representing the total revenue for each data point.
    
    Returns:
    None
    """
    plt.figure(figsize=(10, 6))
    plt.hist(total_revenues, bins=10, color='skyblue', edgecolor='black', alpha=0.7)
    plt.title('Total Revenue Distribution')
    plt.xlabel('Total Revenue')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Example data (you would replace this with your actual data)
total_revenues = [10000, 12000, 9000, 15000, 13000, 11000, 14000, 16000, 17000, 12500, 13500]

# Call the function to plot the histogram
plot_revenue_histogram(total_revenues)
```

## pie chart

```python

def plot_product_sales_pie_chart(product_names, sales_counts):
    """
    Plots a pie chart showing the distribution of sales across different products.
    
    Parameters:
    product_names (list): A list of strings representing the names of products.
    sales_counts (list): A list of integers representing the sales count for each product.
    
    Returns:
    None
    """
    plt.figure(figsize=(8, 8))
    plt.pie(sales_counts, labels=product_names, autopct='%1.1f%%', startangle=140)
    plt.title('Product Sales Distribution')
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.tight_layout()
    plt.show()

# Example data (you would replace this with your actual data)
product_names = ['Product A', 'Product B', 'Product C', 'Product D']
sales_counts = [100, 200, 150, 300]

# Call the function to plot the pie chart
plot_product_sales_pie_chart(product_names, sales_counts)
```

# statistical graph

## heatmap

```python

def plot_heatmap(dataframe):
    """
    Plots a heatmap showing the correlation matrix of numerical columns in the dataset.
    
    Parameters:
    dataframe (DataFrame): Pandas DataFrame containing the dataset.
    
    Returns:
    None
    """
    # Select numerical columns from the dataframe
    numerical_df = dataframe.select_dtypes(include=['float64', 'int64'])
    
    # Calculate the correlation matrix
    correlation_matrix = numerical_df.corr()
    
    # Plot the heatmap
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
    plt.title('Correlation Heatmap of Numerical Columns')
    plt.show()

# Example data (you would replace this with your actual dataset)
data = {
    "Total Revenue": [10000, 12000, 9000, 15000, 13000],
    "Sales count": [100, 200, 150, 300, 250],
    "Product Count": [10, 15, 12, 20, 18]
}

# Create a DataFrame from the example data
df = pd.DataFrame(data)

# Call the function to plot the heatmap
plot_heatmap(df)
```

## corner plot graph 

```python

def plot_pairplot(dataframe):
    """
    Plots a pairplot showing the relationships between pairs of variables in the dataset.
    
    Parameters:
    dataframe (DataFrame): Pandas DataFrame containing the dataset.
    
    Returns:
    None
    """
    plt.figure(figsize=(10, 8))
    sns.pairplot(dataframe)
    plt.title('Pairplot of Variables')
    plt.show()

# Example data (you would replace this with your actual dataset)
data = {
    "Total Revenue": [10000, 12000, 9000, 15000, 13000],
    "Sales count": [100, 200, 150, 300, 250],
    "Product Count": [10, 15, 12, 20, 18]
}

# Create a DataFrame from the example data
df = pd.DataFrame(data)

# Call the function to plot the pairplot
plot_pairplot(df)
```

## violin graph

```python

def plot_violin(dataframe, x_column, y_column):
    """
    Plots a violin plot showing the distribution of a numerical variable across different categories.
    
    Parameters:
    dataframe (DataFrame): Pandas DataFrame containing the dataset.
    x_column (str): The name of the column containing categorical data for the x-axis.
    y_column (str): The name of the column containing numerical data for the y-axis.
    
    Returns:
    None
    """
    plt.figure(figsize=(10, 6))
    sns.violinplot(x=x_column, y=y_column, data=dataframe)
    plt.title(f'Violin Plot of {y_column} Across {x_column}')
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.xticks(rotation=45)
    plt.show()

# Example data (you would replace this with your actual dataset)
data = {
    "Product Name": ['Product A', 'Product B', 'Product A', 'Product B', 'Product A'],
    "Total Revenue": [10000, 12000, 9000, 15000, 13000]
}

# Create a DataFrame from the example data
df = pd.DataFrame(data)

# Call the function to plot the violin plot
plot_violin(df, 'Product Name', 'Total Revenue')
```

# statistical depth 

```python

from scipy.stats import skew, kurtosis

# Assuming you already have your dataset loaded into a DataFrame named df

# 1. Identify numerical columns
numerical_columns = df.select_dtypes(include=[np.number]).columns

# 2. Calculate Mean, Median, Standard Deviation, Skewness, and Kurtosis
statistics = pd.DataFrame({
    'Mean': df[numerical_columns].mean(),
    'Median': df[numerical_columns].median(),
    'Standard Deviation': df[numerical_columns].std(),
    'Skewness': skew(df[numerical_columns]),
    'Kurtosis': kurtosis(df[numerical_columns])
})

# 3. Calculate the Correlation Matrix (only for numerical columns)
correlation_matrix = df[numerical_columns].corr()

# Display the results
print("Statistics:")
print(statistics)
print("\nCorrelation Matrix:")
print(correlation_matrix)

# Additional: Plotting the Correlation Matrix as a Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```
